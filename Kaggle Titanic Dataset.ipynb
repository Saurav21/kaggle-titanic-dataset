```python
def passenger_type(passenger):
    age, sex = passenger
    
    if age <= 16:
        return "child"
    else:
        return sex
```

```python
df["who"] = df[["Age","Sex"]].apply(passenger_type, axis = 1)
```

```python
df["who"]
```

```python
sns.countplot(x = df["Pclass"], hue = df["who"])
```
